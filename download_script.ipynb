```python
import pandas as pd
import os
```

```python
years = [2019, 2020]
months = range(1, 13)
```

```python
for year in years:
    for month in months:
        url = f"https://github.com/DataTalksClub/nyc-tlc-data/releases/download/green/green_tripdata_{year}-{month:02}.csv.gz"
        file_name = f"data/green/green_tripdata_{year}-{month:02}.csv.gz"

        os.system(f"wget {url} -O {file_name}")
```
